```python
class Coord:
    x = 0
    y = 0
    def __str__(self):
        return str(self.x) + "x" + str(self.y)
    def __init__(self,x=0,y=0):
        self.x = x
        self.y = y
    def dist(self):
        return (self.x ** 2 + self.y ** 2) ** 0.5
        
        
```

```python
z = Coord(10,20)  # z is an *instance* of Coord
```

```python
str(z)
```

```python
z.dist()
```

```python
def dictCoord(x=0,y=0):
    me = { 'x':x, 'y':y }
    def dist(d):
        return (d['x'] ** 2 + d['y'] ** 2) ** 0.5
    me['dist'] = dist
    return me
    
    
```

```python
4 ** .5
```

```python
dc = dictCoord(10,20)
```

```python
dc
```

```python
dc['dist']({'x':5,'y':5})
```

```python
dc['dist'](dc)
```
