```python
import seaborn as sns
import pandas as pd
```

```python
om = pd.read_csv('omicron.csv')
```

```python
om.head()
```

```python
om.describe()
```

```python
om['Date'].describe()
```

```python
om['date_parsed'] = pd.to_datetime(om['Date'],format='%m/%d/%Y')
```

```python
om.head()
```

```python
sns.lineplot(data=om,x='date_parsed',y='Percentage of Omicron')
```

```python
sns.lineplot(data=om[om.Country == "United States"],x='date_parsed',y='Percentage of Omicron')
```

```python
sns.lineplot(data=om[om.Country == "Brazil"],x='date_parsed',y='Percentage of Omicron')
```

```python
sns.lineplot(data=om[om.Country.isin(["Brazil","United States"])],x='date_parsed',y='Percentage of Omicron', hue="Country")
```
