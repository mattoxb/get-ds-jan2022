### Barplots

```python
sns.barplot(data=iris)  # What data do you think this is showing us?
```

We can customize this by specifying what we want plotted.

```python
sns.barplot(data=iris, x="species", y="petal_length")
```

**Question:** What would you use this for?

### Boxplots

```python
sns.boxplot(data=iris)
```

```python
sns.boxplot(data=iris, x="species", y="petal_length")
```

What do the box plots tell you that the bar plots don't?  Why would you use these?

### Violin Plots

```python
sns.violinplot(data=iris)
```

```python
sns.violinplot(data=iris, x="species", y="petal_length")
```

What do these kinds of plots show us?

### Swarm Plots

```python
sns.swarmplot(data=iris, x="species", y="petal_length")
```

## Relationship Plots

We can explore relationships between the data.  Interesting stories often happen here!

```python
sns.relplot(data=iris,x='petal_width',y='petal_length')
```

Hm.... interesting.  See that spot on the bottom?

```python
sns.relplot(data=iris,x='petal_width',y='petal_length', hue='species')
```

```python
sns.relplot(data=iris,x='petal_width',y='petal_length',hue='species', kind="line")
```

The above plot is pretty, but it's not clear what it is telling us....

## Pairwise Plots

If you are not sure what could be correlated, these functions can really help!

What do you think the diagonal is for?

```python
sns.pairplot(data=iris,hue='species')
```

### Joint Plots

This requires an x and y.

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue="species", kind="scatter") # Requires Seaborn version 0.11
```
